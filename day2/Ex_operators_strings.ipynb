```python
def is_even(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """
    result = a % 2
    if result == 1:
        return False
    else:
        return True

        
is_even(10)
```

```python
is_even(11)
```

```python
def is_even_2(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """
    if a % 2 == 1:
        return False
    else:
        return True

        
print(is_even_2(10))
print(is_even_2(11))
```

```python
def is_even_3(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """
    if a % 2 == 0:
        return True
        
    return False
        
print(is_even_3(10))
print(is_even_3(11))
```

```python
def is_even_4(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """
    return (a % 2 == 0)  # return True if the result is 0, otherwise returns False
        
print(is_even_4(10))
print(is_even_4(11))
```
