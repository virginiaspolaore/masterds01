```python
# Exercise 3

testo_originale = "Collaboratore"

primi_due = testo_originale[:2]   # "Co"
ultimi_due = testo_originale[-2:] # "re"

nuova_stringa = primi_due + ultimi_due
print(f"Risultato: {nuova_stringa}") # Output: Core
```

```python
parte_iniziale = primi_due.lower()  # "co"
parte_finale = ultimi_due.upper()   # "RE"

risultato_finale = f"{parte_iniziale}{parte_finale}"
print(f"Risultato: {risultato_finale}") # Output: coRE
```
