# Separate features (X) and label (y)

```python
X = iris["data"]
y = iris["target"]

print(f"X shape = {X.shape}\ny shape = {y.shape}")
```

# Split Train and Test datasets

```python
# the proccess is exactly the same we saw for linear regression. 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.15, random_state = 41)
```

```python
[a.shape for a in [X_train, X_test, y_train, y_test]]
```

# Normalize features

```python
# we need to normalize features, because it works better (it's a good practice actually)
# we are preparing the data, not cleaning

from sklearn.preprocessing import MinMaxScaler # it's a technique that allows us to...

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled #.shape
```

```python
y_train #.shape
```

# Training!

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter = 2000) 
```

```python
%%time

logreg.fit(X_train_scaled, y_train) # this fit with logreg, prepared before, receives X_train_scaled, y_train
```

# Evaluation

```python
# real labels
y_test
```

```python
# predicted labels (should be equal)
y_hat = logreg.predict(X_test_scaled)
y_hat

```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_hat, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_hat, average='weighted')
precision
```

```python
recall = recall_score(y_test, y_hat, average='weighted')
recall
```

```python
f1 = f1_score(y_test, y_hat, average='weighted')
f1
```

```python
print(f"F1 Score: {100*f1:.2f}%")
```
