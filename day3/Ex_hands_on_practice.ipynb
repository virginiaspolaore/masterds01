```python
df2 = df2[df2["total (R$)"]<1e4]
```

```python
max_price = df2["total (R$)"].max()
print(f"The maximum prices of the houses is {max_price}")
```

```python
threshold = max_price * 0.9
print(f"The threashold is {threshold}")
```

```python
top_df = df2[df2["total (R$)"]>threshold]
top_df
```

```python
top_df_sorted = top_df.sort_values(
    by = "total (R$)",
    ascending = False)
top_df_sorted.head(20)
```
