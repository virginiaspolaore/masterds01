# Error Handling

```python
try:
    x = 3 / 0

except ZeroDivisionError as e:
    print(e)
```

```python
try:
    x = 10 + "abc"

except TypeError as e:
    print(f"The following error occured: {e}")
```

```python
try:
    my_list = [1,2,3,4,5]
    print(my_list[10])

except Exception as e:
    print(f"The following error type has happened: {e}")
```
